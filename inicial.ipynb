```python
# As colunas com alta cardinalidade como nome,ticket,cabine, nao vao ajudar o modelo pois pode torna-lo menos generalizavel nesse caso vamos eliminar essas colunas
```

```python
# eliminado para a base de treino
treino = treino.drop(["Name","Ticket","Cabin"], axis=1) # nao eliminamos a coluna Id porque iremos usar ela

# e eliminando tambem na base de teste (tudo que fizer na base de treino tem que fazer na base de teste)
teste = teste.drop(["Name","Ticket","Cabin"], axis=1)
```

```python
# Agora precisamos tratar a coluna Age que tem valores vazios tanto no treino quanto no teste: No caso essa coluna não deveria ter valores vazios pois todo mundo tem uma idade.
# como essa coluna se refere a idade dos passageiros podemos fazer uma media de idade para preencher os valores vazios

# verificando a idade media da base de treino
treino.Age.mean()
```

```python
# Agora podemos verificar as colunas com a idade vazia e substituir pela media de idade
treino.loc[treino.Age.isnull(),"Age"] = treino.Age.mean()
# treino.loc:Cria uma máscara booleana True onde Age é nulo.
# isnull(): Seleciona as linhas onde a condição é True e a coluna "Age".
# = mean() substitui os valores pela media da idade

# e fazendo o mesmo para a base de teste
teste.loc[teste.Age.isnull(),"Age"] = teste.Age.mean()
```

```python
# Para a base de treino ainda temos 3 valores vazios na coluna "Embarked"
# Vamos verificar os valores vazios e substituir pela MODA (que é o valor mais frequente no caso dessa coluna é o S)

# verificando os valores na coluna Embarked
treino.Embarked.value_counts()
```

```python
# selecionando a moda dessa coluna
treino.Embarked.mode()[0]
```

```python
# substituindo valores vazios 
treino.loc[treino.Embarked.isnull(),"Embarked"] = treino.Embarked.mode()[0]
```

```python
# para finalizar a base de teste temos um arquivo vazio em Fare, vamos usar a mesma logica de Age e substituir pela moda
# verificando a media da coluna fare da base de teste

teste.loc[teste.Fare.isnull(),"Fare"] = teste.Fare.mean()
```
