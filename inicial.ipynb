```python
# As colunas com alta cardinalidade como nome,ticket,cabine, nao vao ajudar o modelo pois pode torna-lo menos generalizavel nesse caso vamos eliminar essas colunas, pois eloas são colunas object (texto)
```

```python
# eliminado para a base de treino
treino = treino.drop(["Name","Ticket","Cabin"], axis=1) #O parâmetro axis define se a remoção é de linhas ou colunas:
# axis=0 → remove linhas
# axis=1 → remove colunas
# nao eliminamos a coluna PassengerId porque iremos usar ela
```

```python
# E eliminando tambem na base de teste (tudo que fizer na base de treino tem que fazer na base de teste)
teste = teste.drop(["Name","Ticket","Cabin"], axis=1)
```

```python
# Agora precisamos tratar a coluna Age que tem valores vazios tanto no treino quanto no teste: No caso essa coluna não deveria ter valores vazios pois todo mundo tem uma idade.
# como essa coluna se refere a idade dos passageiros podemos fazer uma media de idade para preencher os valores vazios

# verificando a idade media da base de treino
treino.Age.mean()
```

```python
# Agora podemos verificar as colunas com a idade vazia e substituir pela media de idade
treino.loc[treino.Age.isnull(), "Age"] = treino.Age.mean()
# treino.loc:Cria uma máscara booleana True onde Age é nulo.
# isnull(): Seleciona as linhas onde a condição é True e a coluna "Age".
# = mean() substitui os valores pela media da idade
```

```python
# E fazendo o mesmo para a base de teste
teste.loc[teste.Age.isnull(), "Age"] = teste.Age.mean()
```

```python
# Para a base de treino ainda temos 2 valores vazios na coluna "Embarked"
# Vamos verificar os valores vazios e substituir pela MODA (que é o valor mais frequente no caso dessa coluna é o S)

# verificando os valores na coluna Embarked
treino.Embarked.value_counts()
```

```python
# selecionando a moda dessa coluna
treino.Embarked.mode()[0]
```

```python
# substituindo valores vazios 
treino.loc[treino.Embarked.isnull() , "Embarked"] = treino.Embarked.mode()[0]
```

```python
# para finalizar a base de teste temos um arquivo vazio em Fare, vamos usar a mesma logica de Age e substituir pela moda
# verificando a media da coluna fare da base de teste

teste.loc[teste.Fare.isnull(),"Fare"] = teste.Fare.mean()
```

```python
# E para a base de teste selecionamos  apenas valores numericos da base de teste
teste_nr = teste.loc[:, col_teste_nr]
```

```python
# Selecionando um modelo de Classificacao

# Nesse caso vamos selecionar 3 modelos :
# Arvore de Decisao - RandomForest
# Vizinhos Proximos - KNN 
# Regressão Logistica 
```

```python
# Antes de usar esses modelos vamos separar a base em base de treino e validacao

# vamos importar o train_split_test

from sklearn.model_selection import train_test_split
```

```python
# separando a base de treino em x e y
x = treino.drop(["PassengerId","Survived","Sex","Embarked"], axis=1)
y = treino.Survived
```

```python
# separando em treino e validacao
x_train, x_val, y_train, y_val = train_test_split( x, y , test_size=0.33 , random_state=42)
```

```python
# Para a arvore de classificao
# Fazendo a Importacao
from sklearn import tree 
```

```python
# Criando o classificador
clf_ac = tree.DecisionTreeClassifier()
```

```python
# fazendo o fit com os dados
clf_ac = clf_ac.fit(x_train, y_train)
```

```python
# fazendo a previsao
y_pred_ac = clf_ac.predict(x_val)
```

```python
# Para Vizinhos Proximos KNN
# Fazendo a importacao
from sklearn.neighbors import KNeighborsClassifier
```

```python
# criando o classificador
clf_knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# fazendo o fit com os dados
clf_knn = clf_knn.fit(x_train, y_train)
```

```python
# fazendo a previsao
y_pred_knn = clf_knn.predict(x_val)
```

```python
# Para a regressao logistica
#fazendo a importacao
from sklearn.linear_model import LogisticRegression
```

```python
# criando o calssificador
clf_rl = LogisticRegression(random_state=42)
```

```python
# fazendo o fit com os dados
clf_rl = clf_rl.fit(x_train , y_train)
```

```python
# fazendo a previsao
y_pred_rl = clf_rl.predict(x_val)
```

```python
# Vamos fazer a avaliacao dos modelos 

# Accuracia
# Matriz de confusao (ajuda a avaliar a distribiucao de erros)
```

```python
# avaliando a accuracia

# importacao
from sklearn.metrics import accuracy_score
```

```python
# Para a Arvore de Classificacao
accuracy_score(y_val, y_pred_ac)
```

```python
# Para o KNN
accuracy_score(y_val, y_pred_knn)
```

```python
# Para a Regressao Logistica
accuracy_score(y_val, y_pred_rl)
```

```python
# Avaliando a Matriz de Confusao
# importacao
from sklearn.metrics import confusion_matrix
```

```python
# para a arvore de classificacao
confusion_matrix(y_val, y_pred_ac)
```

```python
# para o KNN
confusion_matrix(y_val, y_pred_knn)
```

```python
# para a regressao logistica
confusion_matrix(y_val, y_pred_rl)
```
